```python
import numpy as np
from qiskit import *
from qiskit.providers.aer import QasmSimulator
from qiskit.visualization import plot_histogram
from math import pi
import math
from qiskit import QuantumCircuit
simulator = QasmSimulator()
cir = QuantumCircuit(3,3)
desired_vector = [0,0.1,0.17,0.25,math.sqrt(0.7972),0.25,0.17,0.1]
cir.initialize(desired_vector)
cir.draw()
```

```python
cir.measure([0,1,2], [0,1,2])
cc = transpile(cir, simulator)
job = simulator.run(cc, shots=100000)
result = job.result()
counts = result.get_counts(cc)
plot_histogram(counts)
```
